```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
import time
start=time.time()
p=print(func2(2,4))
end=time.time()
print((end-start)*1000)
#this function calculates power exponentially
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
     result = result + num % 10
     num //= 10
   return result
```

```python
# Sample function call: func3(1234)
import time
start=time.time()
p=print(func3(1234))
end=time.time()
print((end-start)*1000)
#it finds sum of digits of a number
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time
start=time.time()
p=print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
end=time.time()
print((end-start)*1000)
# shows how many elements of list a are present in list b
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
import time
start=time.time()
p=print(func5(16))
end=time.time()
print((end-start)*1000)
#checks whether a number n is perfect square
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
import time
start=time.time()
p=print(func8([3,1,2,7,8,6]))
end=time.time()
print((end-start)*1000)
#sorts array in ascending order using recursive swapping
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
import time
start=time.time()
p=print(func12([1, 2, 3, 4, 5], 2))
end=time.time()
print((end-start)*1000)
#it rotates the array to the right by k position
```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum=0
  for i in arr:
    sum=sum+i
  return sum
```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
print(sum_arr(arr))
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  mini=min(arr)
  return mini
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(array,key):
  count=0
  for i in array :
    if (i==key):
     count=1
     break
  if count==1:
    print(key,"is present")
  else:
    print(key,"is not present")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```
